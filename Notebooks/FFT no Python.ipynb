```
import matplotlib.pyplot as plt
import numpy as np
```

```
n = 1000 # numero de pontos no grafico
tx = 200 # tamanho do eixo x, de 0 a 200
w = 2.0 * np.pi / tx # frequencia angular
```

```
# criando a base de tempo
t = np.linspace(0, tx, n)

# criando os sinais
s1 = 2.0 * np.cos(2.0*w*t)
s2 = 1.0 * np.cos(30.0*w*t)

s = s1 + s2
```

```
freq = np.fft.fftfreq(n)

mascara = freq > 0 # apenas valores positivos
```

```
fft_calculo = np.fft.fft(s)

# para ser tudo positivo, tem que calcular o valor absoluto da transformada
fft_abs = 2.0*np.abs(fft_calculo/n)
```

```
plt.figure(1)
plt.title("Sinal original")
plt.plot(t, s)
```

```
plt.figure(2)
plt.title("Sinal da FFT")
plt.plot(freq[mascara]*n, fft_abs[mascara])
plt.show
```
